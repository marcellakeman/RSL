```python
#Create variable column_names and create a subset
column_names = ['movieName',
                'Metascore_w',
                'Author',
                'AuthorHref',
                'Date',
                'Summary',
                'InteractionsYesCount',
                'InteractionsTotalCount',
                'InteractionsThumbUp',
                'InteractionsThumbDown'
               ]

subset = pd.DataFrame(columns = column_names)
```

```python
#Create a for-loop
for movie in range(len(data)):
    if data.movieName.iloc[movie] == 'interstellar':
        row = data[movie:movie + 1]
        print(row)
        subset.append(row)
```

```python
subset
```

```python
interstellar = data.loc[data['movieName']=='interstellar']
interstellar = interstellar[['movieName', 'Metascore_w', 'Author']]
#interstellar = interstellar[['movieName', 'Metascore_w', 'Author'] & data['Metascore_w'] > interstellar['Metascore_w']]
interstellar.columns = ['movieName', 'Metascore_w', 'Author']
interstellar
```

```python
c_names = ['movieName', 'Author', 'Metascore_w']
```

```python
better_movie = pd.DataFrame(columns = c_names)
better_movie

increased_meta = pd.DataFrame(columns = ['inc_meta'])
increased_meta
```

```python
#Fill in better_movie dataset with rows that contains authors which are identical for 'subset' and 'data' dataset, which metascore
#of the large dataset is higher than the small dataset
for a in range(len(subset)):
    for b in range(len(data)):
        if subset.Author.iloc[a] == data.Author.iloc[b] and subset.Metascore_w.iloc[a] < data.Metascore_w.iloc[b]:
            inc_meta = data.Metascore_w.iloc[b] - subset.Metascore_w.iloc[a] #to calculate the relative increase of metascore
            row = data[c_names].iloc[b]
            better_movie = better_movie.append(row)
            #increased_meta = increased_meta['inc_meta'].append(inc_meta)
            better_movie['inc_meta'] = inc_meta
            print(inc_meta)
```
